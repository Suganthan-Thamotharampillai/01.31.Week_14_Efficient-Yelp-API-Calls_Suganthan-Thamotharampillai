```python
## IMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## LOAD DATASET
# Urls
url_akas="https://datasets.imdbws.com/title.akas.tsv.gz"
url_basics="https://datasets.imdbws.com/title.basics.tsv.gz"
url_ratings="https://datasets.imdbws.com/title.ratings.tsv.gz"

# Dataframes
akas = pd.read_csv(url_akas, sep='\t', low_memory=False)
basics = pd.read_csv(url_basics, sep='\t', low_memory=False)
ratings = pd.read_csv(url_ratings, sep='\t', low_memory=False)
```

<br><br><font size=5><b>Required Preprocessing - Details

<font size=4><b>Title.AKAs
<br><font size=3><b>1. Handling '\N' Placeholder Values

```python
# IDENTIFYING '\N' VALUES
print("\nNumber of '\\N' values are", (akas=='\\N').sum().sum())
(akas=='\\N').sum()
```

```python
# REPLACING '\N' VALUES
akas.replace({'\\N':np.nan}, inplace=True)
print("\nNow the number of '\\N' values are", (akas=='\\N').sum().sum(), "\nAnd the number of 'np.NaN' values are", akas.isna().sum().sum(),'\n\n\\N VALUES\n',26*'-')
print((akas=='\\N').sum(),'\n\nNaN VALUES\n',26*'-')
print(akas.isna().sum())
```

<br><font size=3><b>2. Keep only US Movies

```python
# SETTING REGION to 'US'
akas = akas[akas['region'] == 'US']
akas
```

<br><font size=4><b>Title.BASICS
<br><font size=3><b>1. Handling '\N' Placeholder Values

```python
# IDENTIFYING '\N' VALUES
print("\nNumber of '\\N' values are", (basics=='\\N').sum().sum())
(basics=='\\N').sum()
```

```python
# REPLACING '\N' VALUES
basics.replace({'\\N':np.nan}, inplace=True)
print("\nNow the number of '\\N' values are", (basics=='\\N').sum().sum(), "\nAnd the number of 'np.NaN' values are", basics.isna().sum().sum(),'\n\n\\N VALUES\n',26*'-')
print((basics=='\\N').sum(),'\n\nNaN VALUES\n',26*'-')
print(basics.isna().sum())
```

<br><font size=3><b>2. Eliminate movies that are null for runtimeMinutes

```python
# Dropping movies 'null' for Runtime
print(f"Before dropping, 'null' value for movie runtime is {(basics[(basics['titleType'] == 'movie') & (basics['runtimeMinutes'].isna())].index.shape[0])}");y=basics['titleType'].value_counts()
basics.drop(basics[(basics['titleType'] == 'movie') & (basics['runtimeMinutes'].isna())].index, inplace=True)
print(f"After  dropping, 'null' value for movie runtime is {basics[(basics['titleType'] == 'movie') & (basics['runtimeMinutes'].isna())].index.shape[0]}")
```

```python
# Checking 'movie' value
print(f"\nMOVIE VALUE BEFORE\n{23*'-'}\n{y}\n\nMOVIE VALUE AFTER\n{23*'-'}\n{basics['titleType'].value_counts()}")
```

<br><font size=3><b>3. Eliminate movies that are null for genre

```python
# Dropping movies 'null' for Genre
print(f"\nBefore dropping, 'null' value for movie genre is {(basics[(basics['titleType'] == 'movie') & (basics['genres'].isna())].index.shape[0])}");x=basics['titleType'].value_counts()
basics.drop(basics[(basics['titleType'] == 'movie') & (basics['genres'].isna())].index, inplace=True)
print(f"After  dropping, 'null' value for movie genre is {basics[(basics['titleType'] == 'movie') & (basics['genres'].isna())].index.shape[0]}")
print(f"\nMOVIE VALUE BEFORE\n{23*'-'}\n{x}\n\nMOVIE VALUE AFTER\n{23*'-'}\n{basics['titleType'].value_counts()}")
```

<br><font size=3><b>4. Keep only titleType==Movie

```python
# SETTING DATAFRAME ONLY TO 'movie'
basics = basics[basics['titleType'] == 'movie']
basics
```

<br><font size=3><b>5. Keep startYear 2000-2022

```python
# SETTING DATAFRAME TO startYear 2000-2022
basics = basics[(basics['startYear'] >= '2000') & (basics['startYear'] <= '2022')]
basics
```

<br><font size=3><b>6. Eliminate movies that include "Documentary" in genre

```python
# EXCLUDING DOCUMENTARIES
basics = basics[~(basics['genres'].str.contains('documentary', case=False))]
basics
```

<br><font size=3><b>7. Keep only US movies

```python
# FILTERING BASICS THROUGH AKAS DATAFRAME
basics = basics[(basics['tconst'].isin(akas['titleId']))]
basics
```

<br><font size=4><b>Title.RATINGS
<br><font size=3><b>1. Handling '\N' Placeholder Values

```python
# IDENTIFYING '\N' VALUES
print("\nNumber of '\\N' values are", (ratings=='\\N').sum().sum())
(ratings=='\\N').sum()
```

<br><font size=3><b>2. Keep only US movies

```python
# FILTERING RATINGS THROUGH AKAS DATAFRAME
ratings = ratings[(ratings['tconst'].isin(akas['titleId']))]
ratings
```

<br><br><font size=5><b>Creating Folder & Saving Dataframe

```python
## CREATING FOLDER
# Importing OS
import os

# Create Directory
os.makedirs('Data/', exist_ok=True) 

# Confirm folder created
os.listdir("Data/")
```

```python
## SAVING DATAFRAME as csv.gz
akas.to_csv("Data/title_akas.csv.gz",compression='gzip',index=False)
basics.to_csv("Data/title_basics.csv.gz",compression='gzip',index=False)
ratings.to_csv("Data/title_ratings.csv.gz",compression='gzip',index=False)
```
